# Практика

В ходе практики необходимо проанализировать датасет, $\xi = 1+2$ содержащий информацию о разных видах пингвинов.

Признаки:
* species - тип особи,
* island - остров обитания,
* bill_length_mm - длина клюва,
* bill_depth_mm - высота клюва,
* flipper_length_mm - длина верхних ласт,
* body_mass_g - масса тела,
* sex - пол.


Задачи:
1) Выведите количество строк с пропущенными значениями. Удалите их.
2) Выведите количество дублированных строк. Оставьте только оригинальные строки.
3) Для каждого признака посчитайте среднее, минимум, максимум и количество уникальных значений.
4) Для каждого признака постройте распределение его значений.
